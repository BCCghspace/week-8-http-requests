### Using request parameters

```python
url = "https://api.mapbox.com/styles/v1/mapbox/streets-v11/static/-122.4241,37.78,14.25,0,60/600x600"
resp = requests.get(url)
```

```python
resp.status_code
```

More on 401 status code: <https://developer.mozilla.org/en-US/docs/Web/HTTP/Status/401>

**We're missing an access token**

```python
params = {
    "access_token": "pk.eyJ1IjoiYW5keWVzY2giLCJhIjoiY2puYXFzNTMzMDAxMDNvbnJkMjJ0Nm1xdSJ9.oxRSrlKxJusoV9vRfLLHOA"
}

resp = requests.get(url, params=params)
```

```python
resp.status_code
```

## Census API
